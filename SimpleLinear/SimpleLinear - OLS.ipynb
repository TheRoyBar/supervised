```python
#Brief Theory:
# Let's consider n points in a 2D graph with x and y axes. We draw an imaginary line that represents the regression function.
```

```python
#The line is supposed to be the closest line to each point considering the other points
```

```python
#The distance between each individual point and the line is given by => d(i) = y(i) - y'
# where y(i) is the output of point i, and y' is the mean or average value of y.
# Thus, we can calculate the error function E = sigma(i in range(1,n) y(i) - y'.

# Now since distance can be negative, we take the sum of square of each individual errors to make it E = sigma(i in range(1,n)) y(i) - y')^2
```

```python
# We know that, y = mx+b, where m is the slope and b is the intercept. So, placing this value in the above equation we get:
```

```python
# E = Sigma(i in range(1,n)) (y(i) - mx(i) -b)^2
# Now, in order to minimize this error function, we partially differentiate with respect to m and b.
```

```python
# First, differentiating with b, we get the equation: b = y'-mx'
```

```python
# Then, placing this value in the above equation we get: E = [y(i) -mx(i) - y' +mx']^2
# Now, differentiating this function with respect to m, we get: m = { sigma (x(i)-x')(y(i)-y') }/{sigma (x(i)- x')^2 }
```
